```python
import json
file="./laws_final2.json"
with open(file) as f:
    laws = json.load(f)
```

```python
import glob
import pandas as pd
file_list = glob.glob("/workspace/.gen/Crawler/Law/law_name_list" + "/*")

law_name_list=[]
for idx,fl in enumerate(file_list):
    if idx==0:
        df = pd.read_excel(fl)
        law_name_list.append(df["법령명"].tolist())
law_name_list=sum(law_name_list,[])
```

```python
laws_fin=[]
for lll in laws:
    if lll not in laws_fin:
        laws_fin.append(lll)
    else:print(lll["_source"]["법령명한글"])
```

```python
for idx,lf in enumerate(laws_fin):
    if lf["_source"]["법령명한글"] not in law_name_list: print(lf["_source"]["법령명한글"])
    if lf["_source"]["법령명한글"]=="검사직무대리 운영규정":print(idx)
```

```python
total_res=[]
law_list=[]
for idx,law in enumerate(laws_fin):
    law_name=law["_source"]["법령명한글"]
    print(f"{idx}/{len(laws)} =========== {law_name}")
    law_list.append(law_name)
    contents=[]
    for l in law["_source"]["조문"]:
        # l_num=""
        l_title=""
        l_content=""
        # if "조문번호" in  list(l.keys()):
        #     l_num+=l["조문번호"]
        if "조문내용" in  list(l.keys()):
            l_title+=l["조문내용"]
        if "항" in  list(l.keys()):
            for cont in l["항"]:
                for k,v in cont.items():
                    clause,subclause="",""
                    if k=="호":
                        subclause+= "\n".join(map(lambda x: x["호내용"], v))
                    else:
                        clause+=v
                        clause+="\n"
                    l_content+=clause+subclause
        contents.append(f"{l_title}\n\n{l_content}")
    total_res.append({"law_name":law_name,"contents":contents})
```

# Effective Date,  Promulgation Date, Publication Number

```python
import json
file="./laws_final2.json"
with open(file) as f:
    laws = json.load(f)
laws_fin=[]
for lll in laws:
    if lll not in laws_fin:
        laws_fin.append(lll)
    else:print(lll["_source"]["법령명한글"])
```

```python
total_res=[]
for idx,law in enumerate(laws_fin):
    law_name=law["_source"]["법령명한글"]
    law_name_for_url=law_name.replace(" ","")
    effective_date=law["_source"]["시행일자"]
    publication_number=law["_source"]["공포번호"]
    promulgation_date=law["_source"]["공포일자"]
    base_url=f"https://www.law.go.kr/법령/{law_name_for_url}/({effective_date},{publication_number},{promulgation_date})/"
    print(f"{idx}/{len(laws)} =========== {law_name}")
    contents=[]
    for l in law["_source"]["조문"]:
        clause_number=""
        if "조문번호" in list(l.keys()):
            l_num=l["조문번호"]
            clause_number+=f"제{l_num}조"
        clause_url=base_url+clause_number
        res={"law_name":law_name,"clause_number":clause_number,"url":clause_url}
        total_res.append(res)
```

# Shortened

```python
import json
file="./laws_final2.json"
with open(file) as f:
    laws = json.load(f)
laws_fin=[]
for lll in laws:
    if lll not in laws_fin:
        laws_fin.append(lll)
    else:print(lll["_source"]["법령명한글"])
```

```python
total_res=[]
for idx,law in enumerate(laws_fin):
    law_name=law["_source"]["법령명한글"]
    if "법령약칙명" in list(law["_source"]):
        shortened=law["_source"]["법령약칙명"]
        if shortened is not None:
            res={"law_name":law_name,"shortened":shortened}
            total_res.append(res)
```
