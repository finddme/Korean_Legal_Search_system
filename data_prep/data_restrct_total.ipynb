# Load Data

```python
import json
file="./laws_final2.json"
with open(file) as f:
    laws = json.load(f)
```

```python
import glob
import pandas as pd
file_list = glob.glob("/workspace/.gen/Crawler/Law/law_name_list" + "/*")

law_name_list=[]
for idx,fl in enumerate(file_list):
    if idx==0:
        df = pd.read_excel(fl)
        law_name_list.append(df["법령명"].tolist())
law_name_list=sum(law_name_list,[])
```

```python
laws_fin=[]
for lll in laws:
    if lll not in laws_fin:
        laws_fin.append(lll)
    else:print(lll["_source"]["법령명한글"])
```

# Extract informations

```python
total_res=[]
law_list=[]
for idx,law in enumerate(laws_fin):
    law_name=law["_source"]["법령명한글"]
    if "법령약칙명" in list(law["_source"]):
        shortened=law["_source"]["법령약칙명"]
    else: shortened=""
    law_name_for_url=law_name.replace(" ","")
    effective_date=law["_source"]["시행일자"]
    publication_number=law["_source"]["공포번호"]
    promulgation_date=law["_source"]["공포일자"]
    base_url=f"https://www.law.go.kr/법령/{law_name_for_url}/({effective_date},{publication_number},{promulgation_date})/"
    print(f"{idx}/{len(laws)} =========== {law_name}")
    law_list.append(law_name)
    contents=[]
    for l in law["_source"]["조문"]:
        clause_number=""
        l_num=""
        l_title=""
        if "조문번호" in  list(l.keys()):
            l_num+=l["조문번호"]
            clause_number+=f"제{l_num}조"
        if "조문제목" in list(l.keys()):
            add_cont=l["조문제목"]
            l_title+=f"제{l_num}조({add_cont})"
        else: l_title+=clause_number
        clause_url=base_url+clause_number
        res={"l_title":l_title,"l_content":[],"url":clause_url}
        clauses_format=f"{law_name} {l_title}"
        l_content=""
        if "조문내용" in  list(l.keys()):
            add_cont=l["조문내용"]
            res["l_content"].append(f"{law_name} {add_cont}")
        if "항" in  list(l.keys()):
            for cont in l["항"]:
                for k,v in cont.items():
                    clause,subclause="",""
                    if k=="호":
                        for i in map(lambda x: x["호내용"], v):
                            res["l_content"].append(f"{clauses_format} {i}")
                    else:
                        res["l_content"].append(f"{clauses_format} {v}")
        #             clauses+=clause+subclause
        contents.append(res)
    total_res.append({"law_name":law_name,"shortened":shortened,"contents":contents})
```

# Restruct

```python
total_res2=[]
for idx,tr in enumerate(total_res):
    # if idx==0:
    law_name=tr["law_name"]
    shortened=tr["shortened"]
    for tc in tr["contents"]:
        l_title=tc["l_title"]
        url=tc["url"]
        full_content="\n".join(tc["l_content"])
        for lc in tc["l_content"]:
            data={"law_name":law_name,"shortened":shortened,"law_article":l_title,"law_content":lc, "full_content":full_content,"url":url}
            total_res2.append(data)
```

```python
for tttt in total_res2:
    if "대한민국헌법" in tttt["law_name"]:
        print(tttt)
```
